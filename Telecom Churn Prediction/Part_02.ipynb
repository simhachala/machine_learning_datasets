### Data Preprocessing

#### Splitting categorical columns with unique values less than 2 and more than 2

```python
bin_col = data[cat].nunique()[data[cat].nunique() == 2].keys().tolist()

multi_col = data[cat].nunique()[data[cat].nunique() != 2].keys().tolist()
```

#### Label encoding of binary value columns

```python
encode = LabelEncoder()
for i in bin_col:
    data[i] = encode.fit_transform(data[i])
```

#### Creating dummy variables for mutiple value categorical columns

```python
telecom = pd.get_dummies(data,columns=multi_col,)
```

#### Scaling the data

```python
scale = StandardScaler()
telecom[num] = scale.fit_transform(telecom[num])
```

```python
telecom.shape
```

### Model Building 

#### Splitting of data for modelbulding

```python
x = telecom.drop('Churn',axis=1)
y = telecom['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.3, random_state=0)
```

```python
print("label count before over Sampling 'Yes': {}".format(sum(y_train=='Yes')))
print("label count before over Sampling 'No': {} \n".format(sum(y_train=='No')))

sm = SMOTE(random_state=2)
x_train_res, y_train_res = sm.fit_sample(x_train, y_train.ravel())

print('After OverSampling, the shape of train_X: {}'.format(x_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))

print("label count after over Sampling 'Yes': {}".format(sum(y_train_res=='Yes')))
print("label count after over Sampling 'No': {} \n".format(sum(y_train_res=='No')))
```

### Logistic Regression

```python
logit = LogisticRegressionCV(cv = 10,random_state=0)
model = logit.fit(x_train,y_train)
y_pred = model.predict(x_train)

confusion = pd.DataFrame(confusion_matrix(y_train,y_pred),columns=['Pred_yes','Pred_No'],index=['Act_Yes','Act_No'])
print("Confusion Matrix for Train :")
print("Accuracy for Train :",accuracy_score(y_train,model.predict(x_train)))
confusion
```

```python
confusion = pd.DataFrame(confusion_matrix(y_test,model.predict(x_test)),columns=['Pred_yes','Pred_No'],index=['Act_Yes','Act_No'])
print("Confusion Matrix for Test :")
print("Accuracy for Test :",accuracy_score(y_test,model.predict(x_test)))
confusion
```

### Logistic Regression Model after Over sampling

```python
logit = LogisticRegressionCV(cv = 10,random_state=0)
model = logit.fit(x_train_res,y_train_res)
y_pred = model.predict(x_train_res)

confusion = pd.DataFrame(confusion_matrix(y_train_res,y_pred),columns=['Pred_yes','Pred_No'],index=['Act_Yes','Act_No'])
print("Confusion Matrix for Train  :")
print("Accuracy for Train :",accuracy_score(y_train_res,model.predict(x_train_res)))
confusion
```

```python
confusion = pd.DataFrame(confusion_matrix(y_test,model.predict(x_test)),columns=['Pred_yes','Pred_No'],index=['Act_Yes','Act_No'])
print("Confusion Matrix for Test :")
print("Accuracy for Test :",accuracy_score(y_test,model.predict(x_test)))
confusion
```
